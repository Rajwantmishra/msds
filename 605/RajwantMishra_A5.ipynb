Choose independently two numbers B and C at random from the interval [0,1] with uniform density. Prove that B and C are proper probability distributions. Note that the point (B,C) is then chosen at random in the unit square.


* B and C are proper probability distributions.


A random number generator acting over an interval of numbers (a,b) has a continuous distribution. Since any interval of numbers of equal width has an equal probability of being observed, the curve describing the distribution is a rectangle, with constant height across the interval and 0 height elsewhere. Since the area under the curve must be equal to 1, the length of the interval determines the height of the curve.

Since it's the the uniform random variable B and C are defined on the interval (0,1), the interval has width = 1, the curve has height = 1 over the interval and 0 elsewhere.
$$ f(x) = \left\{\begin{matrix}1 ....if \space 0 \le x \ge 1, \\0....,Otherwise \end{matrix}\right. $$
If E is the event that the B and C falls in the the range of (0,1), then E { B : 0 <= x <= 1 }, and { C : 0 <= x <= 1 }

$$ P(E) = \int_a^b f(x)dx = b-a =>  \int_0^1 f(x)dx = \int_0^1 1dx  =  1-0  = 1   $$

Check 1: The area under the curve of f(x) == 1 <br>
Check 2: For all fx(x) for x>0, we can see that for f(x)  positive everywhere. <br>
Hecne it's a proper probability distributions. <br>

Find the probability that
- a. B+C<1/2

```python
# random numbers from uniform distribution
n = 10000
start = 0
width = 1
data_uniformo = uniform.rvs(size=n, loc = start, scale=width)
data_uniformt = uniform.rvs(size=n, loc = start, scale=width)
data_uniform = data_uniformo + data_uniformt


ax = sns.distplot(data_uniform,
                  bins=1000,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 5,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

We are trying to find A+B < 1/2 , which means that A+B should be any thing between 0 to 1/2. <br>
Here B + C can range from 0 to 2 , as when C and B = 0 (minimum value C + B =0) and when C and B = 1 (Maximum value C + B =2) <br>
The above graph very clearly shows that Sum of A + C lies in the range of 0 and 2. can be written as :<br>
$$f_Z(z) = \begin{cases}
  0, & \text{if } z < 0 \\
  z, & \text{if } 0 \leq z \leq 1 \\    
  2-z, & \text{if } 1 \leq z \leq 2 \\   
  0, & \text{if } 2 < z \\ 
\end{cases}  $$
Let B + C = Z , We will now derive expressions for the cumulative distribution function and the density function of Z.
Now we will find the the area under the curve i.e. we take the integral of the density function f(z) with the limit from 0 to 1/2:

$$ P\Bigg(Z < 1/2\Bigg) =\int_{0}^{1/2} f(z) dz \\
=\left[ \frac{z^2}{2} + constant  \right]_0^\frac{1}{2} \\
=\left[ \frac{\frac{1}{2}^2}{2} + constant  \right] - 0 \cdots doing\space b - a\\
= \left[ \frac{\frac{1}{4}}{2} + constant  \right] - 0
= 1/8 = > .125
$$

 <b> B+C<1/2  = .125 </b>

- b. BC<1/2

```python
# random numbers from uniform distribution
n = 10000
start = 0
width = 1
data_uniformo = uniform.rvs(size=n, loc = start, scale=width)
data_uniformt = uniform.rvs(size=n, loc = start, scale=width)
data_uniform = data_uniformo * data_uniformt
chalf = [data_uniform[i] for i in range(n) if data_uniform[i]<1/2 ]

print("Simulated Probability BC < 1/2 :=>",len(chalf)/10000)

ax = sns.distplot(data_uniform,
                  bins=1000,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 5,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

<b>Probability BC < 1/2 :=> 0.1546</b>

- c. |BC|<1/2


```python
n = 10000
start = 0
width = 1
data_uniformo = uniform.rvs(size=n, loc = start, scale=width)
data_uniformt = uniform.rvs(size=n, loc = start, scale=width)
data_uniform = abs(data_uniformo - data_uniformt)
chalf = [data_uniform[i] for i in range(n) if data_uniform[i]<1/2 ]

print("Simulated Probability |BC|<1/2 :=>",len(chalf)/10000)

ax = sns.distplot(data_uniform,
                  bins=1000,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 5,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

<b>  Probability |BC|<1/2 :=> 0.7429 </b>

- d. max(B,C)<1/2


```python
n = 10000
start = 0
width = 1
data_uniformo = uniform.rvs(size=n, loc = start, scale=width)
data_uniformt = uniform.rvs(size=n, loc = start, scale=width)
data_uniform = np.array([max(data_uniformo[i], data_uniformt[i]) for i in  range(n)])

ax = sns.distplot(data_uniform,
                  bins=1000,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 5,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
chalf = [data_uniform[i] for i in range(n) if data_uniform[i]<1/2 ]

print("Simulated Probability max(B,C)<1/2 :=>",len(chalf)/10000)
```

<b> Probability max(B,C)<1/2 :=> 0.2531 </b>
